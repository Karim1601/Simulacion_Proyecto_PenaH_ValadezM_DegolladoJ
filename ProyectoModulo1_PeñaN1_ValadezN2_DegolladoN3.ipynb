## 1.3 Definición del problema.

## 1.4 Nodos y variables que se decidieron simular y porqué

> * Simulación/estimación de las reservas de los siguientes meses: Al estimar el número de reservaciones podemos darnos cuenta si la situación del hotel se verá favorable o no
> * Simulación/ estimación de los días entre semana que se hospedaron en el hotel: Simular los días de ocupación entre semana es importante porque permite determinar la gestión de recursos, como personal y precios durante esos días. > * Simulación/ estimación de las personas que se pueden hospedar en el hotel: Al simular la cantidad de gente que puede haber en tiempos posteriores puede ayudar al hotel a saber si necesita más personal para atender las necesidades de los huéspedes o permanecen igual
> * Simulación/ estimación de los tiempos de ocupación del hotel: Al igual que las reservas el saber el tiempo de ocupación nos ayudará a saber cómo será la situación del hotel, si tendrán una mayor carga de trabajo o no


## 1.5 Definición de hipótesis y supuestos.

>Hipótesis: Existe una correlación entre las variables simuladas o estimadas (reservas, días entre semana hospedados, cantidad de personas que se pueden hospedar y tiempos de ocupación del hotel), de tal forma que las predicciones en una variable pueden proporcionar información útil para hacer supuestos sobre las otras variables. Esta correlación permite predecir con mayor precisión la situación futura del hotel y tomar decisiones informadas para la gestión de recursos y personal, a fin de satisfacer las necesidades de los huéspedes y mantener una rentabilidad sostenible.
En base a esta hipótesis, se puede suponer que:

>•	Si la simulación o estimación de las reservas muestra un aumento significativo para los próximos meses, es probable que el número de personas que se pueden hospedar en el hotel también aumente, y que la ocupación del hotel en general sea mayor.

>•	Si la simulación o estimación de los días entre semana hospedados muestra un aumento significativo, es probable que la ocupación del hotel en esos días también aumente, lo que podría requerir ajustes en la gestión de recursos y personal.

>•	Si la simulación o estimación de la cantidad de personas que se pueden hospedar muestra un aumento significativo, es probable que las reservas y la ocupación del hotel también aumenten, lo que podría requerir la asignación de más recursos y personal para satisfacer las necesidades de los huéspedes.

>•	Si la simulación o estimación del tiempo de ocupación del hotel muestra un aumento significativo, es probable que el número de reservas y la cantidad de personas que se pueden hospedar también aumenten, lo que podría requerir ajustes en la planificación y asignación de recursos y personal.

>Resumiendo, la hipótesis expandida sostiene que las variables simuladas o estimadas están interconectadas y, por lo tanto, las predicciones en una variable pueden ayudar a hacer supuestos sobre las demás variables. Esto podría permitir una mejor planificación y gestión de los recursos y personal del hotel, lo que a su vez podría mejorar la calidad del servicio y la rentabilidad del negocio.

### Reporte de calidad de datos

```python
def dqr(data):
    #%% Lista de variables de la base de datos
    columns = pd.DataFrame(list(data.columns.values),columns=['Nombres'], index=list(data.columns.values))
    
    #% Lista de tipos de datos
    data_types = pd.DataFrame(data.dtypes, columns=['Data_types'])
    
    #% Lista de valores presentes
    present_values = pd.DataFrame(data.count(), columns=['Present_values'])
    
    #%Lista de valores missing...
    
    
    #% Valores unicos de las columnas
    unique_values = pd.DataFrame(columns=['Unique_values'])
    for col in list(data.columns.values):
        unique_values.loc[col]=[data[col].nunique()]
    
    #% Lista de valores maximos
    max_values= pd.DataFrame(columns=['Max'])
    for col in list(data.columns.values):
        try:
            max_values.loc[col]=[data[col].max()]
        except:
            pass
    
    #% Lista de valores minimos
    min_values= pd.DataFrame(columns=['Min'])
    for col in list(data.columns.values):
        try:
            min_values.loc[col]=[data[col].min()]
        except:
            pass
        
    #% Columna 'categorica' que sea booleana que cuando sea True represente que la variable caracteristica es una var categorica, si es false es numerica
    #% Columna que sea valores unicos de la var categorica, añadir las categorias de registros si la variable es categorica
    
    
    return columns.join(data_types).join(present_values).join(unique_values).join(max_values).join(min_values)
    
```

```python
def kde_scipy(x, x_grid, bandwidth=0.2, **kwargs):
    """Kernel Density Estimation with Scipy"""
    # Note that scipy weights its bandwidth by the covariance of the
    # input data.  To make the results comparable to the other methods,
    # we divide the bandwidth by the sample standard deviation here.
    kde = gaussian_kde(x, bw_method=bandwidth / x.std(ddof=1), **kwargs)
    return kde.evaluate(x_grid)


def kde_statsmodels_u(x, x_grid, bandwidth=0.2, **kwargs):
    """Univariate Kernel Density Estimation with Statsmodels"""
    kde = KDEUnivariate(x)
    kde.fit(bw=bandwidth, **kwargs)
    return kde.evaluate(x_grid)
    
    
def kde_statsmodels_m(x, x_grid, bandwidth=0.2, **kwargs):
    """Multivariate Kernel Density Estimation with Statsmodels"""
    kde = KDEMultivariate(x, 
                          bw='cv_ml',  #bandwidth * np.ones_like(x),
                          var_type='c', **kwargs)
    return kde.pdf(x_grid)


def kde_sklearn(x, x_grid, bandwidth=0.2, **kwargs):
    """Kernel Density Estimation with Scikit-learn"""
    kde_skl = KernelDensity(bandwidth=bandwidth, **kwargs)
    kde_skl.fit(x[:, np.newaxis])
    # score_samples() returns the log-likelihood of the samples
    log_pdf = kde_skl.score_samples(x_grid[:, np.newaxis])
    return np.exp(log_pdf)


kde_funcs = [kde_statsmodels_u, kde_statsmodels_m, kde_scipy, kde_sklearn]
kde_funcnames = ['Statsmodels-U', 'Statsmodels-M', 'Scipy', 'Scikit-learn']
```


#### Prueba de Hipótesis



```python
# Asignar un número a cada día de la semana
dias_num = np.arange(7)

# Hipótesis nula: los días de la semana son igualmente probables
# Generar 1000 muestras aleatorias bajo la hipótesis nula
n = 10000
null_samples = np.random.choice(dias_num, size=(n, len(dias_num)))

# Estadístico de prueba: diferencia entre la frecuencia observada y la esperada
observed_freqs = np.histogram(dias_num, bins=np.arange(8))[0]
expected_freqs = np.ones_like(observed_freqs) * len(null_samples) / len(dias_num)
test_statistic = np.sum((observed_freqs - expected_freqs)**2 / expected_freqs)

# Valor p: probabilidad de obtener un estadístico de prueba al menos tan extremo como el observado bajo la hipótesis nula
null_test_stats = np.sum((np.histogram(sample, bins=np.arange(8))[0] - expected_freqs)**2 / expected_freqs for sample in null_samples)
p_value = np.mean(null_test_stats >= test_statistic)

# Graficar el histograma de los datos y las muestras simuladas bajo la hipótesis nula
fig, ax = plt.subplots()
ax.hist(dias_num, bins=np.arange(8)-0.5, fc='gray', histtype='stepfilled', alpha=0.3, label='Datos observados')
ax.hist(null_samples.flatten(), bins=np.arange(8)-0.5, fc='gray', histtype='step', alpha=0.5, label='Muestras bajo H0')
ax.set_xticks(np.arange(7))
ax.set_xticklabels(dias_semana)
ax.legend()
ax.set_title('Histograma de los días de la semana')
plt.show()

# Imprimir el estadístico de prueba y el valor p
print('Estadístico de prueba:', test_statistic)
print('Valor p:', p_value)


# interpretar el resultado
alpha = 0.05

if p_value < alpha:
    print("Se rechaza la hipótesis nula.")
else:
    print("No se puede rechazar la hipótesis nula.")
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Lista de días de la semana
dias_semana = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']

# Simulación de los días de la semana
simulacion = [random.choice(dias_semana) for i in range(1000)]

# Cálculo de la función de distribución empírica
def funcion_distribucion_empirica(dia, simulacion):
    n = len(simulacion)
    return sum([1 for i in simulacion if i <= dia])/n

# Cálculo de la función de distribución empírica para cada día
funciones_distribucion_empirica = [funcion_distribucion_empirica(dia, simulacion) for dia in dias_semana]

# Cálculo de la función de distribución teórica
funcion_distribucion_teorica = np.linspace(0, 1, num=7)

# Cálculo de la diferencia máxima entre las funciones de distribución empírica y teórica
max_diferencia = max(abs(funciones_distribucion_empirica - funcion_distribucion_teorica))

# Gráfico de las funciones de distribución empírica y teórica
plt.plot(dias_semana, funciones_distribucion_empirica, label='Función de Distribución Empírica')
plt.plot(dias_semana, funcion_distribucion_teorica, label='Función de Distribución Teórica')
plt.title('Comparación entre Funciones de Distribución Empírica y Teórica')
plt.xlabel('Día de la Semana')
plt.ylabel('Probabilidad')
plt.legend()
plt.show()

# Resultado del test de Kolmogorov
if max_diferencia > 1.36/np.sqrt(1000):
    print('Se rechaza la hipótesis de que la distribución empírica es igual a la teórica')
else:
    print('No se puede rechazar la hipótesis de que la distribución empírica es igual a la teórica')
```

En este bloque, se realizó una prueba de hipótesis para determinar si la media poblacional de la cantidad de personas que se pueden hospedar en el hotel es igual a 2.
```python
def aceptacion_rechazo(data, n_samples):
    x_min, x_max = data.min(), data.max()
    x_kde, y_kde = sns.kdeplot(data, bw_adjust=0.25).get_lines()[0].get_data()
    y_max = y_kde.max()
    plt.close()

    samples = []
    while len(samples) < n_samples:
        x_rand = np.random.uniform(x_min, x_max)
        y_rand = np.random.uniform(0, y_max)
        if y_rand < np.interp(x_rand, x_kde, y_kde):
            samples.append(x_rand)
    return samples

n_samples = 1000

for column in ['children', 'babies', 'adults']:
    samples = aceptacion_rechazo(data[column], n_samples)
    plt.hist(samples, bins=20, alpha=0.5, label='Muestras generadas')
    sns.histplot(data[column], kde=True, bins=20, label='Datos originales')
    plt.title(f"Muestras generadas usando Aceptación y Rechazo para {column.capitalize()}")
    plt.legend()
    plt.show()
```

Se  generan muestras de datos a partir de una distribución de probabilidad usando el método de aceptación y rechazo. Además, se compara la distribución de las muestras generadas con los datos originales para evaluar qué tan bien se ajusta la distribución generada a los datos reales.

```python
# Montecarlo
def montecarlo_simulation(data, n_simulations):
    estimations = []
    for _ in range(n_simulations):
        sample = data.sample(frac=1, replace=True)
        estimations.append(sample.mean())
    return estimations

n_simulations = 1000
montecarlo_results = montecarlo_simulation(data['total_personas'], n_simulations)

plt.hist(montecarlo_results, bins=20, alpha=0.5, label='Montecarlo Simulations')
plt.title("Montecarlo Simulations for Average Number of Guests")
plt.xlabel("Average Number of Guests")
plt.ylabel("Frequency")
plt.legend()
plt.show()

montecarlo_mean = np.mean(montecarlo_results)
montecarlo_std = np.std(montecarlo_results)
print(f"Montecarlo Mean: {montecarlo_mean}, Montecarlo Standard Deviation: {montecarlo_std}")
```

```python
datos = np.array(lead_time_array)

# Calcular la varianza original
var_original = np.var(datos)

# Implementar el método de reducción de varianza
n = 10000
m = len(datos) // n
bloques = np.split(datos[:m*n], m)
medias_bloques = np.array([np.mean(bloque) for bloque in bloques])
var_medias_bloques = np.var(medias_bloques)
var_reducida = var_medias_bloques + (1 - 1/n) * var_original

# Calcular el intervalo de confianza al 95%
z = stats.norm.ppf(0.975)
delta = z * np.sqrt(var_reducida / len(bloques))
intervalo_inf = np.mean(medias_bloques) - delta
intervalo_sup = np.mean(medias_bloques) + delta

# Graficar los datos y las medias de los bloques
fig, ax = plt.subplots()
ax.plot(datos, '.', alpha=0.3, label='Datos')
ax.plot(np.repeat(np.arange(m)*n + n/2, 2), np.tile(medias_bloques, 2), '-', linewidth=2, label='Medias de los bloques')
ax.axhline(np.mean(medias_bloques), color='r', linestyle='--', label='Media de las medias de los bloques')
ax.fill_betweenx(ax.get_ylim(), intervalo_inf, intervalo_sup, alpha=0.2, color='g', label='Intervalo de confianza al 95%')
ax.legend()
ax.set_xlabel('Índice')
ax.set_ylabel('Valor')
ax.set_title('Método de reducción de varianza')
plt.show()

# Imprimir la varianza original, la reducción de varianza media y el intervalo de confianza al 95%
print('Varianza original:', var_original)
print('Varianza reducida:', var_reducida)
print('Intervalo de confianza al 95%:', intervalo_inf, intervalo_sup)

```
