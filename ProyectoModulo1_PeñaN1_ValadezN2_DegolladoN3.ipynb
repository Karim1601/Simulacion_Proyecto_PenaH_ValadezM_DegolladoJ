## 1.5 Definición de hipótesis y supuestos.

>Hipótesis: Existe una correlación entre las variables simuladas o estimadas (reservas, días entre semana hospedados, cantidad de personas que se pueden hospedar y tiempos de ocupación del hotel), de tal forma que las predicciones en una variable pueden proporcionar información útil para hacer supuestos sobre las otras variables. Esta correlación permite predecir con mayor precisión la situación futura del hotel y tomar decisiones informadas para la gestión de recursos y personal, a fin de satisfacer las necesidades de los huéspedes y mantener una rentabilidad sostenible.
En base a esta hipótesis, se puede suponer que:

>•	Si la simulación o estimación de las reservas muestra un aumento significativo para los próximos meses, es probable que el número de personas que se pueden hospedar en el hotel también aumente, y que la ocupación del hotel en general sea mayor.

>•	Si la simulación o estimación de los días entre semana hospedados muestra un aumento significativo, es probable que la ocupación del hotel en esos días también aumente, lo que podría requerir ajustes en la gestión de recursos y personal.

>•	Si la simulación o estimación de la cantidad de personas que se pueden hospedar muestra un aumento significativo, es probable que las reservas y la ocupación del hotel también aumenten, lo que podría requerir la asignación de más recursos y personal para satisfacer las necesidades de los huéspedes.

>•	Si la simulación o estimación del tiempo de ocupación del hotel muestra un aumento significativo, es probable que el número de reservas y la cantidad de personas que se pueden hospedar también aumenten, lo que podría requerir ajustes en la planificación y asignación de recursos y personal.

>Resumiendo, la hipótesis expandida sostiene que las variables simuladas o estimadas están interconectadas y, por lo tanto, las predicciones en una variable pueden ayudar a hacer supuestos sobre las demás variables. Esto podría permitir una mejor planificación y gestión de los recursos y personal del hotel, lo que a su vez podría mejorar la calidad del servicio y la rentabilidad del negocio.

### Reporte de calidad de datos

```python
def dqr(data):
    #%% Lista de variables de la base de datos
    columns = pd.DataFrame(list(data.columns.values),columns=['Nombres'], index=list(data.columns.values))
    
    #% Lista de tipos de datos
    data_types = pd.DataFrame(data.dtypes, columns=['Data_types'])
    
    #% Lista de valores presentes
    present_values = pd.DataFrame(data.count(), columns=['Present_values'])
    
    #%Lista de valores missing...
    
    
    #% Valores unicos de las columnas
    unique_values = pd.DataFrame(columns=['Unique_values'])
    for col in list(data.columns.values):
        unique_values.loc[col]=[data[col].nunique()]
    
    #% Lista de valores maximos
    max_values= pd.DataFrame(columns=['Max'])
    for col in list(data.columns.values):
        try:
            max_values.loc[col]=[data[col].max()]
        except:
            pass
    
    #% Lista de valores minimos
    min_values= pd.DataFrame(columns=['Min'])
    for col in list(data.columns.values):
        try:
            min_values.loc[col]=[data[col].min()]
        except:
            pass
        
    #% Columna 'categorica' que sea booleana que cuando sea True represente que la variable caracteristica es una var categorica, si es false es numerica
    #% Columna que sea valores unicos de la var categorica, añadir las categorias de registros si la variable es categorica
    
    
    return columns.join(data_types).join(present_values).join(unique_values).join(max_values).join(min_values)
    
```

```python
def kde_scipy(x, x_grid, bandwidth=0.2, **kwargs):
    """Kernel Density Estimation with Scipy"""
    # Note that scipy weights its bandwidth by the covariance of the
    # input data.  To make the results comparable to the other methods,
    # we divide the bandwidth by the sample standard deviation here.
    kde = gaussian_kde(x, bw_method=bandwidth / x.std(ddof=1), **kwargs)
    return kde.evaluate(x_grid)


def kde_statsmodels_u(x, x_grid, bandwidth=0.2, **kwargs):
    """Univariate Kernel Density Estimation with Statsmodels"""
    kde = KDEUnivariate(x)
    kde.fit(bw=bandwidth, **kwargs)
    return kde.evaluate(x_grid)
    
    
def kde_statsmodels_m(x, x_grid, bandwidth=0.2, **kwargs):
    """Multivariate Kernel Density Estimation with Statsmodels"""
    kde = KDEMultivariate(x, 
                          bw='cv_ml',  #bandwidth * np.ones_like(x),
                          var_type='c', **kwargs)
    return kde.pdf(x_grid)


def kde_sklearn(x, x_grid, bandwidth=0.2, **kwargs):
    """Kernel Density Estimation with Scikit-learn"""
    kde_skl = KernelDensity(bandwidth=bandwidth, **kwargs)
    kde_skl.fit(x[:, np.newaxis])
    # score_samples() returns the log-likelihood of the samples
    log_pdf = kde_skl.score_samples(x_grid[:, np.newaxis])
    return np.exp(log_pdf)


kde_funcs = [kde_statsmodels_u, kde_statsmodels_m, kde_scipy, kde_sklearn]
kde_funcnames = ['Statsmodels-U', 'Statsmodels-M', 'Scipy', 'Scikit-learn']

```

```python
def aceptacion_rechazo(data, n_samples):
    x_min, x_max = data.min(), data.max()
    x_kde, y_kde = sns.kdeplot(data, bw_adjust=0.25).get_lines()[0].get_data()
    y_max = y_kde.max()
    plt.close()

    samples = []
    while len(samples) < n_samples:
        x_rand = np.random.uniform(x_min, x_max)
        y_rand = np.random.uniform(0, y_max)
        if y_rand < np.interp(x_rand, x_kde, y_kde):
            samples.append(x_rand)
    return samples

n_samples = 1000

for column in ['children', 'babies', 'adults']:
    samples = aceptacion_rechazo(data[column], n_samples)
    plt.hist(samples, bins=20, alpha=0.5, label='Muestras generadas')
    sns.histplot(data[column], kde=True, bins=20, label='Datos originales')
    plt.title(f"Muestras generadas usando Aceptación y Rechazo para {column.capitalize()}")
    plt.legend()
    plt.show()
```

```python
# Montecarlo
def montecarlo_simulation(data, n_simulations):
    estimations = []
    for _ in range(n_simulations):
        sample = data.sample(frac=1, replace=True)
        estimations.append(sample.mean())
    return estimations

n_simulations = 1000
montecarlo_results = montecarlo_simulation(data['total_personas'], n_simulations)

plt.hist(montecarlo_results, bins=20, alpha=0.5, label='Montecarlo Simulations')
plt.title("Montecarlo Simulations for Average Number of Guests")
plt.xlabel("Average Number of Guests")
plt.ylabel("Frequency")
plt.legend()
plt.show()

montecarlo_mean = np.mean(montecarlo_results)
montecarlo_std = np.std(montecarlo_results)
print(f"Montecarlo Mean: {montecarlo_mean}, Montecarlo Standard Deviation: {montecarlo_std}")
```

```python
def kde_scipy(x, x_grid, bandwidth=0.2, **kwargs):
    """Kernel Density Estimation with Scipy"""
    # Note that scipy weights its bandwidth by the covariance of the
    # input data.  To make the results comparable to the other methods,
    # we divide the bandwidth by the sample standard deviation here.
    kde = gaussian_kde(x, bw_method=bandwidth / x.std(ddof=1), **kwargs)
    return kde.evaluate(x_grid)


def kde_statsmodels_u(x, x_grid, bandwidth=0.2, **kwargs):
    """Univariate Kernel Density Estimation with Statsmodels"""
    kde = KDEUnivariate(x)
    kde.fit(bw=bandwidth, **kwargs)
    return kde.evaluate(x_grid)
    
    
def kde_statsmodels_m(x, x_grid, bandwidth=0.2, **kwargs):
    """Multivariate Kernel Density Estimation with Statsmodels"""
    kde = KDEMultivariate(x, 
                          bw='cv_ml',  #bandwidth * np.ones_like(x),
                          var_type='c', **kwargs)
    return kde.pdf(x_grid)


def kde_sklearn(x, x_grid, bandwidth=0.2, **kwargs):
    """Kernel Density Estimation with Scikit-learn"""
    kde_skl = KernelDensity(bandwidth=bandwidth, **kwargs)
    kde_skl.fit(x[:, np.newaxis])
    # score_samples() returns the log-likelihood of the samples
    log_pdf = kde_skl.score_samples(x_grid[:, np.newaxis])
    return np.exp(log_pdf)


kde_funcs = [kde_statsmodels_u, kde_statsmodels_m, kde_scipy, kde_sklearn]
kde_funcnames = ['Statsmodels-U', 'Statsmodels-M', 'Scipy', 'Scikit-learn']

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Creamos un array con datos aleatorios
datos = lead_time_array

# Definimos la función de densidad de kernel
kde = gaussian_kde(datos)

# Generamos puntos para graficar la PDF
x = np.linspace(datos.min(), datos.max(), 100)

# Calculamos la PDF en los puntos generados
pdf = kde(x)

# Graficamos la PDF obtenida a partir de la función de densidad de kernel
plt.plot(x, pdf, '-')
plt.xlabel('Datos')
plt.ylabel('PDF(x)')
plt.show()
```
